```python
import random
from collections import Counter
import matplotlib.pyplot as plt

def roll_dice():
    return random.randint(1, 6)

def simulate_dice_rolls(n):
    results = [roll_dice() for _ in range(n)]
    return results

def analyze_results(results):
    counts = Counter(results)
    probabilities = {number: count / len(results) for number, count in counts.items()}
    return counts, probabilities

def visualize_results(counts):
    plt.bar(counts.keys(), counts.values())
    plt.xlabel('Dice Number')
    plt.ylabel('Frequency')
    plt.title('Dice Roll Distribution')
    plt.show()

# Number of times the dice will be rolled
num_rolls = 10000

# Simulate dice rolls
results = simulate_dice_rolls(num_rolls)

# Analyze the results
counts, probabilities = analyze_results(results)

# Print the results
print(f"Roll Counts: {counts}")
print(f"Probabilities: {probabilities}")

# Visualize the results
visualize_results(counts)
```
